```python
import json; 
import os;
```

```python
messages= json.load(open("discord.json"))
messages[0]
```

```python
messages[1]['attachments'][0]['url']
```

```python
from tqdm import tqdm;
```

```python
urls= []
i=0
for message in tqdm(messages):
    try: 
        url_i= message['attachments'][0]['url']
        idx_i= url_i.index("?ex=")
        urls.append(url_i[:idx_i])
    except Exception as e:
        print(e)
        i+=1
        
print("\nFailed Images= ", i)
        
        
```

```python
len(urls)
```

```python
with open("urls.json", "w") as f:
    json.dump(urls, f)
```
